## Dropping 4 Columns

```python
train.drop(['UsageClass', 'CheckoutType' , 'CheckoutYear' , 'CheckoutMonth'], axis=1, inplace = True)
```

```python
def normalize_text(s):
    s = s.lower()
    
    # remove punctuation that is not word-internal (e.g., hyphens, apostrophes)
    s = re.sub('\s\W',' ',s)
    s = re.sub('\W\s',' ',s)
    
    # make sure we didn't introduce any double spaces
    s = re.sub('\s+',' ',s)
    
    return s

train['Title_new'] = [normalize_text(s) for s in train['Title']]
train['Subject_new'] = [normalize_text(s) for s in train['Subjects']]
```

```python
train.drop(['Title' , 'Subjects' ], axis = 1, inplace = True)
```

```python
output_dict = {
    "BOOK" : 1,
    "SOUNDDISC" : 2,
    "VIDEOCASS" : 3,
    "VIDEODISC" : 4,
    "SOUNDCASS" : 5,
    "MUSIC" : 6,
    "CR" : 7,
    "MIXED" :8
}

train['MaterialTypeName'] = train['MaterialType'].apply(lambda x: output_dict[x])
#test['MaterialTypeName'] = test['MaterialType'].apply(lambda x: output_dict[x])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
train.drop(['MaterialType'], axis = 1, inplace = True)
```

```python
X = np.array(train.ix[:, train.columns != 'MaterialTypeName'])
y = np.array(train.ix[:, train.columns == 'MaterialTypeName'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
print("Before OverSampling, counts of label 'BOOK': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label 'SOUNDDISC': {}".format(sum(y_train==2)))
print("Before OverSampling, counts of label 'VIDEOCASS': {}".format(sum(y_train==3)))
print("Before OverSampling, counts of label 'SOUNDCASS': {}".format(sum(y_train==4)))
print("Before OverSampling, counts of label 'MUSIC': {} ".format(sum(y_train==5)))
print("Before OverSampling, counts of label 'CR': {}".format(sum(y_train==6)))
print("Before OverSampling, counts of label 'MIXED': {}".format(sum(y_train==7)))

sm = SMOTE(random_state=2)
X_train_res, y_train_res = sm.fit_sample(X_train, y_train.ravel())
```

```python
print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label 'BOOK': {}".format(sum(y_train_res=='BOOK')))
print("After OverSampling, counts of label 'SOUNDDISC': {}".format(sum(y_train_res=='SOUNDDISC')))
print("After OverSampling, counts of label 'VIDEOCASS': {}".format(sum(y_train_res=='VIDEOCASS')))
print("After OverSampling, counts of label 'SOUNDCASS': {}".format(sum(y_train_res=='SOUNDCASS')))
print("After OverSampling, counts of label 'MUSIC': {} ".format(sum(y_train_res=='MUSIC')))
print("After OverSampling, counts of label 'CR': {}".format(sum(y_train_res=='CR')))
print("After OverSampling, counts of label 'MIXED': {}".format(sum(y_train_res=='MIXED')))
```
